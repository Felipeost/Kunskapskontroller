### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# Plotting can be done either with the object-oriented style or the pyplot method.
# The object oriented style is recommended in general, particularly for parts of larger projects. 
# However, the pyplot style can be very convenient for quick interactive work.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure is the whole thing, can be seen as a canvas. It keeps track of axes, artists etc.
# Artists is everything visible on the figure. Texts, axes, axis, lines. 
# Axes is an Artist attached to a Figure that contains a region for plotting data.
# Axis objects set the scale and limits and generate ticks and ticklabels.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# The expected input data type is numpy arrays, or objects that can be put in to the numpy.asarray to get an numpy array (ex list).
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4)
y= x**2
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("X-axis") 
ax.set_ylabel("Y-axis")  
ax.set_title("First Plot") 
plt.show()
```

```python
x = np.linspace(-4, 4)
y= x**2
plt.plot(x, y)
plt.xlabel("X-axis") 
plt.ylabel("Y-axis")  
plt.title("First Plot") 
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
fruit_colors = ["purple", "green", "orange", "yellow", "#00FF00"]
bar_labels = ["Grapes", "Apples", "Oranges", "Lemons", "Limes"]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 10))

ax1.scatter(random_data_x, random_data_y, color = "red")
ax1.set_title("SCATTER PLOT", fontsize = 30)
ax1.set_xlabel("X-axis", fontsize = 20)
ax1.set_ylabel("Y-axis", fontsize = 20)
ax1.annotate('Minimum y-value', xy=(-0.9, -3.1796043723464056), xytext = (-3, -2.5), fontsize = 15,
            arrowprops=dict(facecolor='blue', shrink = 0.05))

ax2.bar(names, values, label = bar_labels, color = fruit_colors)
ax2.set_title("FRUIT BAR", fontsize = 30)
ax2.set_xlabel("Fruits", fontsize = 20)
ax2.set_ylabel("Amount", fontsize = 20)
ax2.legend(bar_labels, title = "Fruit Color")

plt.show()
```
